```python
# Coding Task B — Arc Statistics with spaCy
import random
random.seed(42)

# Use sentences from Brown corpus or generate a small corpus
nltk.download("brown", quiet=True)
from nltk.corpus import brown

brown_sents = [" ".join(s) for s in brown.sents()[:500]]
sample_corpus = random.sample(brown_sents, 50)

dep_counts = Counter()
distances = []
left_arcs = 0
right_arcs = 0

for sent in sample_corpus:
    doc = spacy_nlp(sent)
    for tok in doc:
        if tok.dep_ != "ROOT":
            dep_counts[tok.dep_] += 1
            dist = abs(tok.i - tok.head.i)
            distances.append(dist)
            if tok.i < tok.head.i:
                left_arcs += 1
            else:
                right_arcs += 1

print("Top 10 dependency relations:")
for dep, count in dep_counts.most_common(10):
    print(f"  {dep:<15} {count}")

print(f"\nAverage dependency distance: {sum(distances)/len(distances):.2f}")
total_arcs = left_arcs + right_arcs
print(f"Left arcs:  {left_arcs}/{total_arcs} ({100*left_arcs/total_arcs:.1f}%)")
print(f"Right arcs: {right_arcs}/{total_arcs} ({100*right_arcs/total_arcs:.1f}%)")
```
